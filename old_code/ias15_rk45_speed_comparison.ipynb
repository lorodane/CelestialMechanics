```python
import numpy as np
import rebound
import tools.sitnikov_integrator as sit
import random
import time
from scipy.integrate import solve_ivp
```

```python
e = 0.5
sit.set_eps(e)
```

```python
T = 500
```

```python
# Generate list of random initial values
N_samp = 10

t_random = np.array([2*np.pi*random.random() for _ in range(N_samp)])
v_random = np.array([2.5*random.random() for _ in range(N_samp)])
```

```python

times_rk45 = np.zeros((N_samp,))

for i in range(N_samp):
    t0 = t_random[i]
    z0 = 0
    v0 = v_random[i]

    start = time.time()

    ans = solve_ivp(fun = sit.phase_velocity,
                    y0 = np.array([z0, v0]),
                    t_span = (t0, t0 + T),
                    rtol = 1e-9,
                    atol = 1e-12)
    
    end = time.time()
    times_rk45[i] = end-start
```
