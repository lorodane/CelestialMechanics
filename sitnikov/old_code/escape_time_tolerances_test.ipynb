# Imports — integrator, NumPy and plotting tools


```python
import tools.sitnikov_integrator as sit
import numpy as np
import matplotlib.pyplot as plt
```

# Set model parameter `eps`


```python
e = 0.5
sit.set_eps(e)
```

# Define `test_plot_tolerances(rtol, atol)` — compute escape times and plot


```python
def test_plot_tolerances(rtol, atol):    
    t0 = 0
    v_arr = np.linspace(2.5, 3, 200)

    t_esc_arr = np.zeros(np.size(v_arr))

    Tmax = 600

    for i in range(v_arr.size):
        t_esc_arr[i] = sit.escape_time(t0, v_arr[i], Tmax, rtol = rtol, atol = atol)

    inf_mask = np.isinf(t_esc_arr)
    t_esc_arr[inf_mask] = np.full(inf_mask.sum(), Tmax)

    plt.plot(v_arr, t_esc_arr)
    plt.plot(v_arr[inf_mask], np.full(inf_mask.sum(), Tmax), 'k.')
```

# Run `test_plot_tolerances` for several tolerance pairs

```python
test_plot_tolerances(1e-3, 1e-6)
test_plot_tolerances(1e-4, 1e-7)
test_plot_tolerances(1e-5, 1e-8)
plt.legend(["rtol =1e-3, atol = 1e-6", "",
            "rtol = 1e-4, atol = 1e-7", "",
            "rtol = 1e-6, atol = 1e-8"], "")
```
